```python
# Notebook overview: 
# This notebook builds the API used for class NoiseModel in simulator_final. 
# NoiseModel monitors simulator noise via SDE d(lambda_t) and Radon-Nikodym (RN)

# Class: NoiseModel(a, b, a_tilde, T, dt)
# SDE: d(lambda_t) = a*d(lambda_t)dt + b*d(lambda_t)dW_t
# Methods: 
# - sample_path(lambda0): 
#       Returns (times, lam, dW) arrays
# - rn_weight(lam, dW): 
#       Computes RN weight using drift difference and Girsanov theorem

# Example usage: 
a = lambda lam: 0.0
b = lambda lam: 0.1
a_tilde = lambda lam: 0.05
noise = NoiseModel(a, b, a_tilde, T = 1.0, dt = 0.01)
times, lam, dW = noise.sample_path(lambda0 = 0.0)
w = noise.rn_weight(lam, dW)
```
